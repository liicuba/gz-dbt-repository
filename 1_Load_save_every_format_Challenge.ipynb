## Dealing with json files

Execute the following cell to initialize a dictionary `d`



```python
d = {"firstname": "leonardo",
     "lastname": "dicaprio",
     "job": "actor"}

print(d)
```

*Create* a file named `actors.json` from `d`. Have a look at the lecture if you are lost with the syntax.


```python
import json

#dictionary
d = {
    "firstname": "leonardo",
    "lastname": "dicaprio",
    "job": "actor"
}

#json path
file_path = '/content/actors.json'

#create a json file
with open(file_path, 'w') as f:
    json.dump(d, f)

#printing a friendly message
print(f'Arquivo JSON "{file_path}" criado com sucesso.')
```

Open the new `actors.json` file and add a new key `nationality` (Leonardo Dicaprio is American). \
Save the result to the same file name. Check your file.


```python
#your code here

#json file path
file_path = '/content/actors.json'

#open the json file and load it into a dic
with open (file_path, 'r') as f:
  data  = json.load(f)

#add the new key
data ['nationality'] = 'American'

#write themodified dictionary back to the json file
with open (file_path, 'w') as f:
  json.dump(data, f)

#check your file
print(f'JSON file "{file_path}" updated successfully.')
```

Add in each dataframe a new `restaurant_id`column (id = 1 for restaurant_1 and id = 2 for restaurant_2) and concatenate the two dataframes into a single one named `df_sales`.  

Pay attention to the `axis`along which you will concatenate the two dataframes.

```python
# your code here

import pandas as pd

# Path to your Excel file
file_path = '/content/restaurants_accountability.xlsx'

# Use ExcelFile to access specific sheets with pandas library
excel_file = pd.ExcelFile(file_path)

# Access sheet names
sheet_name_1 = excel_file.sheet_names[0]  # first sheet for restaurant 1
sheet_name_2 = excel_file.sheet_names[1]  # second sheet for restaurant 2

# Load data for both sheets
df_sales_restaurant_1 = pd.read_excel(excel_file, sheet_name=sheet_name_1)
df_sales_restaurant_2 = pd.read_excel(excel_file, sheet_name=sheet_name_2, skiprows=3, usecols='c:f') #os comandos skiproows e usecols so usam as colunas e linhas que tem info

# Drop empty rows for both dataframes
df_sales_restaurant_1.dropna(axis=0, how='all', inplace=True)
df_sales_restaurant_2.dropna(axis=0, how='all', inplace=True)

# Add restaurant_id column
df_sales_restaurant_1['restaurant_id'] = 1
df_sales_restaurant_2['restaurant_id'] = 2

# Concatenate the two dataframes
df_sales = pd.concat([df_sales_restaurant_1, df_sales_restaurant_2], axis=0)

#ja printa no dataframe
df_sales
```

Export the concatenated dataframe to a new `restaurants_accountability_full.xlsx` file in a sheet called `All`



```python
# your code here

# Export the concatenated dataframe to a new Excel file
output_file_path = '/content/restaurants_accountability_full.xlsx'
df_sales.to_excel(output_file_path, sheet_name='All', index=False)

# Display the concatenated dataframe
print("Concatenated DataFrame exported successfully to 'restaurants_accountability_full.xlsx' in sheet 'All'.")
df_sales
```

## Working with an unidentified format


Someone from the team found a file named [`secret.pkl`](https://drive.google.com/file/d/1_oia73SoCZtZFsGbDrsffkSfdaj2Gt5O/view?usp=sharing).

Can you help them open it and find what's inside? Stackoverflow might have some answers!

```python
# your code here
import pickle

# Path to the pickle file
pickle_file_path = '/content/secret.pkl'

# Load the pickle file ('rb' open a file in binary read mode)
with open(pickle_file_path, 'rb') as file:
    data = pickle.load(file)

# Display the type of the loaded data
print(f"Type of the loaded data: {type(data)}")

# Display the contents of the data
if isinstance(data, dict):
    for key, value in data.items():
        print(f"{key}: {value}")
elif isinstance(data, list):
    for item in data:
        print(item)
elif isinstance(data, pd.DataFrame):
    print(data.head())
else:
    print(data)
```

Note: if you want to open your file into a dataframe, you can usually just use pandas!
